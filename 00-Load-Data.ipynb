```python
import mysql.connector
import os
import csv
```

```python
host = "localhost"
user = "chris"
password = "password"
database = "midas"

# Establish a database connection
connection = mysql.connector.connect(
    host=host,
    user=user,
    password=password,
    database=database
)

cursor = connection.cursor(buffered=True)
```

```python
directory = 'Data/Raw/midas/flow'
files = os.listdir(directory)
csv_files = [file for file in files if file.endswith(".csv")]
```

```python
#Drop table
drop_table_sql = "DROP TABLE IF EXISTS {}".format('full_data')
# Execute the SQL statement to drop the table
cursor.execute(drop_table_sql)
# Commit the changes (if using InnoDB engine)
connection.commit()
```

```python
#Create table
create_table_sql = """
CREATE TABLE IF NOT EXISTS full_data (
    yr int,
    data_date date,
    data_hour	int,
    time_period	int,
    site_ID	VARCHAR(255),
    Lane	VARCHAR(255),
    Short_Veh	int,
    Long_Veh	int,
    Total_Flow	int,
    S_Flag	int,
    L_Flag	int,
    T_Flag	int,
    Latitude	float,
    Longitude	float,
    Network_Link	int,
    NTIS_Version	float,
    S_No_of_mins	int,
    L_No_of_mins	int,
    T_No_of_mins	int,
    DayOfWeek	int,
    yearnum	int,
    weeknum	int,
    daynum	int,
    HolidayPeriod	int,
    UK_Holiday	VARCHAR(255),
    V_Flag	int,
    Source VARCHAR(255)

)
"""

# Execute the SQL statement to create the table
cursor.execute(create_table_sql)
```

```python
cursor.execute("SELECT * FROM full_data") # insert data into the full_data table
field_names = [i[0] for i in cursor.description]
field_names = "("  + ','.join(field_names) + ")"

files_processed = 0

for next_csv in csv_files:
    
    with open('Data/Raw/midas/flow/{}'.format(next_csv), 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        
        # Skip the header row if it exists
        next(csv_reader, None)
        
        # Iterate through the rows in the CSV file
        for row in csv_reader:
            # Define the INSERT statement based on your table structure
            # Assuming the CSV columns match the table columns in order
            insert_sql = "INSERT INTO {} {} VALUES ({})".format('full_data',field_names,('%s,' * 27)[:-1])
            
            # Execute the INSERT statement with data from the CSV
            cursor.execute(insert_sql, (row[0], row[1], row[2],row[3],row[4],row[5],row[6],row[7],row[8],row[9],row[10],row[11],row[12],row[13],row[14],row[15],row[16],row[17],row[18],row[19],row[20],row[21],row[22],row[23],row[24],row[25],row[26]))
    
    # Commit the changes (if using InnoDB engine)
    connection.commit()

    files_processed += 1
    
    print('{} files processed, out of {}'.format(files_processed,len(csv_files)))
```
